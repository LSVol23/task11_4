```python
def car_price_regression(filename: str):
    import pandas as pd
    import numpy as np
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import root_mean_squared_error

    # 1. Загрузка данных
    df = pd.read_csv(filename)

    # 2. Выбор признаков
    features = ["Year", "Engine HP", "Engine Cylinders",
                "highway MPG", "city mpg", "Popularity"]
    target = "MSRP"

    X = df[features]
    y = df[target]

    # 3. Обработка пропусков
    X = X.fillna(X.mean())

    # 4. Разделение на train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    # 5. Создание и обучение модели
    model = LinearRegression()
    model.fit(X_train, y_train)

    # 6. Предсказания
    y_pred = model.predict(X_test)

    # 7. RMSE линейной регрессии
    rmse_lr = root_mean_squared_error(y_test, y_pred)

    # 8. Предсказание среднего значения из train
    mean_train = y_train.mean()
    y_pred_dummy = np.full(len(y_test), mean_train)

    # 9. Корень из глупой модели
    rmse_dummy = root_mean_squared_error(y_test, y_pred_dummy)

    # 10. Сравнение
    difference = rmse_dummy - rmse_lr

    print(f"RMSE линейной регрессии: {rmse_lr}")
    print(f"RMSE глупой модели: {rmse_dummy}")
    print(f"Разница (dummy - linear): {difference}")

    return rmse_lr, rmse_dummy, difference
```
